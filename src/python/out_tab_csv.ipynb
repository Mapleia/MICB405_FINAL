```python
import pandas as pd
import os
```

```python
star_dir_path = "../PRJNA576974_star"

# 1. Get all the folder names in the directory
star_dir_list = os.listdir(star_dir_path)
all_df = pd.DataFrame()

if len(star_dir_list) == 0:
    print("No folders found in the directory")
else:

    for proj_dir in star_dir_list:
        print(proj_dir)
        # 2. Get all the files in the folder
        proj_dir_path = os.path.join(star_dir_path, proj_dir)
        proj_file_list = os.listdir(proj_dir_path)
        if len(proj_file_list) == 0:
            print("No files found in the folder")
        else:
            # 3. Get the file name and file path
            # get files with the extension .tab
            tab_files = list(filter(lambda x: x.endswith(".tab"), proj_file_list))

            for file in tab_files:
                file_path = os.path.join(proj_dir_path, file)
                # 4. Read the file
                cols = ['GENE_ID', 'UNSTRANDED_COUNT', 'COUNT_1ST_READ', 'COUNT_2ND_READ']
                df = pd.read_csv(file_path, sep="\t", header=None, skiprows=4, names=cols)
                df['BioProject'] = proj_dir

                srr_id = file.split("_")[1]
                df['SRR_ID'] = srr_id
                all_df = pd.concat([all_df, df], ignore_index=True)

# order by the count cols
all_df = all_df.sort_values(by=['UNSTRANDED_COUNT', 'COUNT_1ST_READ', 'COUNT_2ND_READ'], ascending=False)
```
